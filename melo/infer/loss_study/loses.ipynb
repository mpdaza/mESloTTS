```python
import pandas as pd
```

```python
df = pd.read_csv('loss_silero.csv', index_col=0)
df 
```

```python
df['G'] = df.index * 200
df['G']
```

```python
# Filtered data  by 1000 to 1000 G's
df_filtered = df[df['G'] % 1000 == 0]
df_filtered
```

FEMALE

```python
female = df_filtered.sort_values(by='female', ascending=True)
col1 = female.iloc[:, [0]].reset_index(drop=True)
col2 = female.iloc[:, [2]].reset_index(drop=True) 
# female = pd.concat([female.iloc[:, [0]], female.iloc[:, [2]].reset_index(drop=True)], axis=1)
female = pd.concat([col1,col2], axis=1)
female
```

```python
female.describe()
```

```python
female.head(100)
```

```python
female[female['female'] > 36]
```

```python
female.shape
```

```python
residuals = female[female['female'] < 36]
bins = [33, 34, 35, 36]  # Limites de los rangos
labels = ['33-34', '34-35', '35-36']  # Nombres para los rangos
residuals['rango'] = pd.cut(residuals['female'], bins=bins, labels=labels, right=False)
residuals.rango.value_counts()
```

```python
residuals[residuals['rango'] == '33-34']
```

```python
residuals[residuals['rango'] == '34-35']
```

MALE

```python
male = df_filtered.sort_values(by='male', ascending=True)
col1 = male.iloc[:, [1]].reset_index(drop=True) 
col2 = male.iloc[:, [2]].reset_index(drop=True) 
male = pd.concat([col1, col2], axis=1)
male
```
